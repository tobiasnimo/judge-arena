```python
from datasets import load_dataset
import pandas as pd

ds = load_dataset("lmsys/mt_bench_human_judgments")
df = ds['human'].to_pandas()
```

```python
df.drop(df[df['turn'] == 2].index, inplace=True)
sample = df.sample(n=100).reset_index(drop=True)
sample
```

```python
dataset = []

for index, row in sample.iterrows():
    conversation_a = list(row['conversation_a'])
    answer_a = dict(conversation_a[1]).get('content')
    conversation_b = list(row['conversation_b'])
    answer_b = dict(conversation_b[1]).get('content')
    question = dict(conversation_a[0]).get('content')
    winner = row['winner']

    _ = {
        "question": question,
        "model_a": answer_a,
        "model_b": answer_b,
        "winner": winner
    }
    dataset.append(_)
```

```python
import json

with open("../../datasets/bestof.json", "w") as f:
    json.dump(dataset, f)
```
